```python
import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
from pandas import read_csv
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
import keras
```

```python
# Train data
file_train = 'ADNI_Normalized_Train.csv'
file_test = 'ADNI_Normalized_Test.csv'
df_train = read_csv(file_train,header=0)
df_test = read_csv(file_test,header=0)
```

```python
all_col = ['RID', 'VISCODE', 'DX', 'PTGENDER', 'PTEDUCAT', 'PTETHCAT', 'PTRACCAT',
       'APOE4', 'CDRSB', 'ADAS11', 'ADAS13', 'ADASQ4', 'MMSE',
       'RAVLT.immediate', 'RAVLT.learning', 'RAVLT.forgetting',
       'RAVLT.perc.forgetting', 'LDELTOTAL', 'TRABSCOR', 'FAQ', 'MOCA',
       'EcogPtMem', 'EcogPtLang', 'EcogPtVisspat', 'EcogPtPlan', 'EcogPtOrgan',
       'EcogPtDivatt', 'EcogPtTotal', 'EcogSPMem', 'EcogSPLang',
       'EcogSPVisspat', 'EcogSPPlan', 'EcogSPOrgan', 'EcogSPDivatt',
       'EcogSPTotal', 'IMAGEUID', 'Ventricles', 'Hippocampus', 'WholeBrain',
       'Entorhinal', 'Fusiform', 'MidTemp', 'ICV', 'AGE']
```

```python
used_col = ['CDRSB', 'ADAS11', 'ADAS13', 'ADASQ4', 'MMSE',
       'RAVLT.immediate', 'RAVLT.learning', 'RAVLT.forgetting',
       'RAVLT.perc.forgetting', 'LDELTOTAL', 'TRABSCOR', 'FAQ', 'MOCA',
       'EcogPtMem', 'EcogPtLang', 'EcogPtVisspat', 'EcogPtPlan', 'EcogPtOrgan',
       'EcogPtDivatt', 'EcogPtTotal', 'EcogSPMem', 'EcogSPLang',
       'EcogSPVisspat', 'EcogSPPlan', 'EcogSPOrgan', 'EcogSPDivatt',
       'EcogSPTotal', 'IMAGEUID', 'Ventricles', 'Hippocampus', 'WholeBrain',
       'Entorhinal', 'Fusiform', 'MidTemp', 'ICV', 'AGE']
```

```python
train_label = df_train['DX']
test_label = df_test['DX']

y_train = np.array(train_label)
y_test = np.array(test_label)

for i in range(len(y_train)):
    if y_train[i] == 'Dementia':
        y_train[i] = 1
    else:
        y_train[i] = 0
        
for i in range(len(y_test)):
    if y_test[i] == 'Dementia':
        y_test[i] = 1
    else:
        y_test[i] = 0
        
y_train = y_train.astype('int32')
y_test = y_test.astype('int32')
```

```python
train_data = df_train[used_col]
test_data = df_test[used_col]

X_train = train_data.values
X_test = test_data.values
```

```python
print('Train data', X_train.shape)
print('Train label', y_train.shape)
print('Test data', X_test.shape)
print('Test label', y_test.shape)
```

```python
# define the keras model
model = Sequential()
model.add(Dense(16, input_dim=36, activation='relu'))
model.add(Dense(8, activation='relu', name='my_intermediate_layer'))
model.add(Dense(1, activation='sigmoid'))
```

```python
# compile the keras model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# fit the keras model on the dataset
model.fit(X_train, y_train, epochs=150, batch_size=10)
```

```python
loss, accuracy = model.evaluate(X_test, y_test)
print('Accuracy: %.2f' % (accuracy*100))
print('Loss: %.2f' % (loss))
```

```python
# extract the new embedding
feature_extractor = keras.Model(
   inputs=model.inputs,
   outputs=model.get_layer(name="my_intermediate_layer").output
)
train_extracted_features = feature_extractor(X_train)
new_train_rep = (train_extracted_features.numpy())[:,:]
test_extracted_features = feature_extractor(X_test)
new_test_rep = (test_extracted_features.numpy())[:,:]
new_train_rep.shape
```

```python
new_train = df_train[['RID', 'VISCODE', 'DX']]
new_test = df_test[['RID', 'VISCODE', 'DX']]
```

```python
new_train = new_train.values
new_test = new_test.values
```

```python
new_train = np.concatenate((new_train, new_train_rep), axis=1)
new_test = np.concatenate((new_test, new_test_rep), axis=1)
```

```python
new_train_df = pd.DataFrame(data = new_train, columns = ['RID', 'VISCODE', 'DX', 'F1', 'F2', 'F3', 'F4', 'F5', 'F6', 'F7', 'F8'])
new_test_df = pd.DataFrame(data = new_test, columns = ['RID', 'VISCODE', 'DX', 'F1', 'F2', 'F3', 'F4', 'F5', 'F6', 'F7', 'F8'])
```
