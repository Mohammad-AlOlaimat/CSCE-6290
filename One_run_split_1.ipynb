```python
 # to panelize positive (converter) misclassification
def binary_cross_entropy(y, yhat):
    alpha = 0.7
    loss = -(tf.math.reduce_mean((alpha * y * tf.math.log(yhat + 1e-6)) + ((1.0- alpha) * (1 - y) * tf.math.log(1 - yhat + 1e-6)), axis=-1))
    
    return loss
```

```python
# Create a BiGRU model
def create_BiGRU_model(drout, L2):
    batch_shape = (None, time_steps, num_features_in_each_time_step)
    model = Sequential()
    model.add(Masking(-1, batch_input_shape=batch_shape))
    model.add(Bidirectional(GRU(32, input_shape=(None, num_features_in_each_time_step), activity_regularizer=l2(L2), return_sequences=True)))# activation = 'relu'
    model.add(Dropout(drout))
    model.add(Bidirectional(GRU(16, activity_regularizer=l2(L2), return_sequences=True, name='my_intermediate_layer')))#activation = 'tanh'
    model.add(Dropout(drout))
    model.add(Dense(1,  activation='sigmoid'))
    
    model.compile(loss= binary_cross_entropy, optimizer='adam',metrics=['accuracy'])#mean_absolute_error binary_crossentropy
    return model
```

```python
# fbeata_function method to calculate f2 score
def overall_fbeta_function(pred, actual):
    
    # reshape the output
    if len(pred.shape) > 2:
        pred = np.reshape(pred, (pred.shape[0], pred.shape[1]*pred.shape[2]))
        actual = np.reshape(actual, (actual.shape[0], actual.shape[1]*actual.shape[2]))
    
    y = []
    yhat = []
    
    for i in range(pred.shape[0]):
        for j in range(pred.shape[1]):
            if pred[i,j] > 0.5:
                pred[i,j] = 1
            else:
                pred[i,j] = 0 
    
    for i in range(pred.shape[0]):
        COUNTER = 0
        while (COUNTER < actual.shape[1]):
            if actual[i,COUNTER] != -1:
                COUNTER+=1
            else:
                break
        for j in range(COUNTER):
            y.append(actual[i,j])
            yhat.append(pred[i,j])
    
    return fbeta_score(y, yhat, beta=2)
```

```python
def do_ablation(data, target, Tdata, Ttarget, iteration):
    X_train = data
    y_train = target

    X_test = Tdata
    y_test = Ttarget
    
    #without demographic
    drout = 0.4
    L2 = 0.001
    epochs_ = 70
    batch_size_ = 8
    model = create_BiGRU_model(drout, L2)
    history = model.fit(X_train, y_train, epochs=epochs_, batch_size = batch_size_, shuffle = True)
    
    #train
    train_loss, train_acc = model.evaluate(X_train, y_train, batch_size = batch_size_)
    train_pred = model.predict(X_train)
    overall_fbeta_function(train_pred, y_train)
    
    #test
    score, acc = model.evaluate(X_test, y_test, batch_size = batch_size_)
    pred = model.predict(X_test)
    
    # reshape the output
    pred = np.reshape(pred, (pred.shape[0], pred.shape[1]*pred.shape[2]))
    real = np.reshape(y_test, (y_test.shape[0], y_test.shape[1]*y_test.shape[2]))
    new_df = pd.DataFrame(pred)
    new_df2 = pd.DataFrame(real)
    prediction = new_df
    prediction = prediction.values
    
    for i in range(prediction.shape[0]):
        for j in range(prediction.shape[1]):
            if prediction[i,j] > 0.5:
                prediction[i,j] = 1
            else:
                prediction[i,j] = 0
    
    ground_T = new_df2
    ground_T = ground_T.values
    
    predicted_l = np.zeros((len(prediction)))
    real_l = np.zeros((len(ground_T)))
    dx = prediction.shape[1] - 1
    for i in range(len(prediction)):
        predicted_l[i] = prediction[i,dx]
    for i in range(len(ground_T)):
        real_l[i] = ground_T[i,dx]
    CM = confusion_matrix(real_l, predicted_l, labels=[0,1])
    
    sensitivity = CM[1,1] / (CM[1,1] + CM[1,0])
    specificity = CM[0,0] / (CM[0,0] + CM[0,1])
    
    # Table of results
    col = 'Iteration '#+str(iteration)
    metrics_results_df = pd.DataFrame(columns = [col])

    metrics_results_df.loc[len(metrics_results_df)] = [round(accuracy_score(ground_T[:, -1], prediction[:, -1]), 3)]
    metrics_results_df.loc[len(metrics_results_df)] = [round(roc_auc_score(ground_T[:, -1], prediction[:, -1]), 3)]
    metrics_results_df.loc[len(metrics_results_df)] = [round(fbeta_score(ground_T[:, -1], prediction[:, -1], beta=2), 3)] 
    metrics_results_df.loc[len(metrics_results_df)] = [round(sensitivity, 3)] 
    metrics_results_df.loc[len(metrics_results_df)] = [round(specificity, 3)]  
    metrics_results_df.loc[len(metrics_results_df)] = [round(acc, 3)]
    metrics_results_df.loc[len(metrics_results_df)] = [round(train_acc, 3)]
    metrics_results_df.loc[len(metrics_results_df)] = [round(score, 3)]
    metrics_results_df.loc[len(metrics_results_df)] = [round(train_loss, 3)]
    metrics_results_df.loc[len(metrics_results_df)] = [round(overall_fbeta_function(prediction, ground_T), 3)]
    metrics_results_df.loc[len(metrics_results_df)] = [round(overall_fbeta_function(train_pred, y_train), 3)]
    
    return metrics_results_df
```

```python
# Table of results
metrics_results_df = pd.DataFrame(columns = ['Metrics'])
metrics_results_df.loc[len(metrics_results_df)] = ['Accuracy (Test)']
metrics_results_df.loc[len(metrics_results_df)] = ['ROC_AUC (Test)']
metrics_results_df.loc[len(metrics_results_df)] = ['F-2 (Test)'] 
metrics_results_df.loc[len(metrics_results_df)] = ['Sensitivity (Test)'] 
metrics_results_df.loc[len(metrics_results_df)] = ['Specificity (Test)']  
metrics_results_df.loc[len(metrics_results_df)] = ['Overall accuracy (Test)']
metrics_results_df.loc[len(metrics_results_df)] = ['Overall accuracy (Train)']
metrics_results_df.loc[len(metrics_results_df)] = ['Overall loss (Test)']
metrics_results_df.loc[len(metrics_results_df)] = ['Overall loss (Train)']
metrics_results_df.loc[len(metrics_results_df)] = ['Overall F2 (Test)']
metrics_results_df.loc[len(metrics_results_df)] = ['Overall F2 (Train)']

time_s = int(input("Please enter time points:\n"))
for i in range(5):
    data, target, Tdata, Ttarget = prepare_data(time_s)
    result = do_ablation(data, target, Tdata, Ttarget, i+1)
    metrics_results_df = pd.concat([metrics_results_df, result], axis=1)
metrics_results_df.to_csv('metrics_results_df_split_1.csv', index = False)
```
