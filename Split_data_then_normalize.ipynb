```python
# Normalization for Train
# To apply z score for each feature separatly
from scipy import stats

columns_name = list(train_dataset.columns)
for i in range(4, len(columns_name)):
    if i == 5 or i == 6 or i == 7: # skip normalization
        continue
    else:
        # Z_SCORE
        
        #train_dataset.iloc[:, i] = stats.zscore(train_dataset.iloc[:, i])
        
        
        # min-max normalization
        
        temp_data = train_dataset.iloc[:,i]
        max_value = temp_data.max()
        min_value = temp_data.min()
        for j in range(len(train_dataset)):
            train_dataset.iat[j, i] = (train_dataset.iloc[j, i]-min_value)/(max_value - min_value)
```

```python
# Normalization for Test
# To apply z score for each feature separatly
from scipy import stats

columns_name = list(test_dataset.columns)
for i in range(4, len(columns_name)):
    if i == 5 or i == 6 or i == 7: # skip normalization
        continue
    else:
        # Z_SCORE
        
        #test_dataset.iloc[:, i] = stats.zscore(test_dataset.iloc[:, i])
        
        
        # min-max normalization
        
        temp_data = test_dataset.iloc[:,i]
        max_value = temp_data.max()
        min_value = temp_data.min()
        for j in range(len(test_dataset)):
            test_dataset.iat[j, i] = (test_dataset.iloc[j, i]-min_value)/(max_value - min_value)
```
