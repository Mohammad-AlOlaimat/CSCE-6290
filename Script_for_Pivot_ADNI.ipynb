```python
def converter_vs_noconverter(df):
    #num_of_features = 18
    columns_name = list(df.columns)
    # to apply one hot encoding            
    temp_data = df.values
    new_dataset = np.empty((temp_data.shape[0], temp_data.shape[1]))# two extra columns for each time point
    for i in range(new_dataset.shape[0]):
        for j in range(temp_data.shape[1]):
            if (str(temp_data[i, j]) == 'MCI' or str(temp_data[i, j]) == 'Dementia'):
                if str(temp_data[i, j]) == 'MCI':
                    new_dataset[i, j] = 0
                if str(temp_data[i, j]) == 'Dementia':
                    new_dataset[i, j] = 1
            else:
                new_dataset[i, j] = temp_data[i, j]
    
    new_df = pd.DataFrame(new_dataset, columns = columns_name)

    return new_df


df1 = converter_vs_noconverter(df1)
df2 = converter_vs_noconverter(df2)
df3 = converter_vs_noconverter(df3)
df4 = converter_vs_noconverter(df4)
df5 = converter_vs_noconverter(df5)
df6 = converter_vs_noconverter(df6)
df7 = converter_vs_noconverter(df7)
df8 = converter_vs_noconverter(df8)
df9 = converter_vs_noconverter(df9)
df10 = converter_vs_noconverter(df10)
df11 = converter_vs_noconverter(df11)
df12 = converter_vs_noconverter(df12)
df13 = converter_vs_noconverter(df13)
df14 = converter_vs_noconverter(df14)
df15 = converter_vs_noconverter(df15)
df16 = converter_vs_noconverter(df16)
df17 = converter_vs_noconverter(df17)
df18 = converter_vs_noconverter(df18)
df19 = converter_vs_noconverter(df19)
df20 = converter_vs_noconverter(df20)
df21 = converter_vs_noconverter(df21)
#df22 = converter_vs_noconverter(df22)
```
